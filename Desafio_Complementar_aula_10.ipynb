```python
import pandas as pd
```

```python
companhias = pd.read_excel("best-selling game consoles.xlsx",sheet_name="consoles")

companhias.head()
```

```python
# Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias.

companhias = companhias.groupby('Company')[['Company']].count()

companhias
```

```python
# Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente.

companhias.column = ['qtd']
companhias = (companhias
              .sort_values('qtd',ascending=False)
              .reset_index())
companhias
```

```python
# Sobre o dataframe anterior, adicione a coluna "Market Value" presente na tabela do arquivo "company.txt".

market_value = pd.read_csv("company.txt")

companhias_market_value = companhias.merge(market_value, on='Company', how='left')

companhias_market_value
```

```python
# Qual o valor mdio de Market Value por console?

media_console = companhias_market_value['Market Value'].sum() / companhias_market_value['qtd'].sum()

media_console
```

```python
companhias_market_value['Media Consola'] = round(companhias_market_value['Market Value'] / companhias_market_value['qtd'],2)

companhias_market_value
```
